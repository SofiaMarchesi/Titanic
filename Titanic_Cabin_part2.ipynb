```
cabin_test = test1['Cabin']
```

```
print(len(cabin1_test))
print(len(cabin1_train))
```

```
cabin_dummie_train = pd.get_dummies(cabin1_train, prefix='Cabin')
cabin_dummie_train.head(10)
```

There is no Cabin_T in test dataset, so will put zeros in this feature.

```
import numpy as np

Cabin_T = pd.DataFrame({'Cabin_T': np.zeros(cabin1_test.shape[0], dtype=np.int8)})
cabin_dummie_test = pd.get_dummies(cabin1_test, prefix='Cabin')
cabin_dummie_test = pd.concat([cabin_dummie_test,Cabin_T], axis=1)
cabin_dummie_test.head(10)
```

```
test2.head(10)
test3 = pd.concat([test2,cabin_dummie_test], axis=1)
test3_ = test3.drop('Unnamed: 0', axis=1)
test3_.head(10)
```
